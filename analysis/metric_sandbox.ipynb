```python
# Add parent directory to Python path for imports
import sys
sys.path.append('..')

import pandas as pd
import numpy as np

from metrics.smape import smape
from config import TARGET_COL

df = pd.read_csv("../data/train.csv")

y_true = df[TARGET_COL].values

IMAGE_DIR = "/content/drive/MyDrive/amazon_ml/images"
```

```python
mean_pred = np.full_like(y_true, y_true.mean())
median_pred = np.full_like(y_true, np.median(y_true))
log_mean_pred = np.expm1(np.mean(np.log1p(y_true)))
log_mean_pred = np.full_like(y_true, log_mean_pred)

print("Mean prediction SMAPE:", smape(y_true, mean_pred))
print("Median prediction SMAPE:", smape(y_true, median_pred))
print("Log-mean prediction SMAPE:", smape(y_true, log_mean_pred))
```

```python
df["price_bucket"] = pd.qcut(df[TARGET_COL], q=5, labels=False)

for bucket in sorted(df["price_bucket"].unique()):
    mask = df["price_bucket"] == bucket
    score = smape(
        y_true[mask],
        median_pred[mask]
    )
    print(f"Bucket {bucket} SMAPE: {score:.2f}")
```
